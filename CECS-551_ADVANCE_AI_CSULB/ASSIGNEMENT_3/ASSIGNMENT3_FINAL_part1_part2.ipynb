```python
import pandas as pd
import numpy as np
```

```python
class NN:
    def __init__(self,input_size,layers,output_size):
        brain = {}
        brain['W1'] = np.random.randn(input_size,layers[0]) 
        brain['b1'] = np.zeros((1,layers[0]))
        brain['W2'] = np.random.randn(layers[0],output_size)
        brain['b2'] = np.zeros((1,output_size))  
        self.brain = brain
    def forward(self,x):
        W1=self.brain['W1']
        W2=self.brain['W2']
        b1=self.brain['b1']
        b2=self.brain['b2']
        z1=np.dot(x,W1) + b1
        a1=np.tanh(z1) 
        z2=np.dot(a1,W2)+b2
        y_= sigmoid(z2)
        self.cache = (a1,y_)
        return y_
    def backward(self,x,y,learning_rate=0.001):
        W1=self.brain['W1']
        W2=self.brain['W2']
        a1,y_ = self.cache
        d2 = y_ - y
        dw2 = np.dot(a1.T,d2) 
        db2 = np.sum(d2,axis=0)
        d1 = (1-np.square(a1))*np.dot(d2,W2.T) 
        dw1 = np.dot(x.T,d1)  
        db1 = np.sum(d1,axis=0)
        self.brain['W1'] -= learning_rate*dw1
        self.brain['W2'] -= learning_rate*dw2
        self.brain['b1'] -= learning_rate*db1
        self.brain['b2'] -= learning_rate*db2
    def yes(self,x):
        y_out = self.forward(x)
        return y_out
        
```

```python
def sigmoid(x):
    xx = 1/(1+np.exp(-x))
    return xx
```

```python
def loss(y,p):
    l = -(np.mean(y*np.log(p) + (1-y)*np.log(1-p)))
    print(l)
    return l   
```

```python
def train(X,Y,model,epochs,logs=True):
    for i in range(epochs):
        yfinal = model.forward(X)
        l = loss(Y,yfinal)
        model.backward(X,Y)
        if(i%50==0):
            print("Epoch %d Loss %.4f"%(i,l),end='\n')
```

```python
model = NN(input_size=2,layers=[2],output_size=1)
```
